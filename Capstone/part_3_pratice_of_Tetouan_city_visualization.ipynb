```python
import pandas as pd
from datetime import datetime
d_parser = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')
tetouan= pd.read_csv("Tetuan City power consumption.csv", parse_dates=['DateTime'], date_parser=d_parser)
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

half_year_locator = mdates.MonthLocator(interval=6)
ax.xaxis.set_major_locator(half_year_locator) # Locator for major axis only.

ax.plot(tetouan['DateTime'], tetouan['Zone 1 Power Consumption'])
plt.xlabel=("Time")
plt.ylabel=("Power Consumption")
plt.title=("Zone1 Power Consumption")
```

```python
tetouan.index=tetouan['DateTime']
tetouan
```

```python
tetouan_grouped=tetouan.resample('M').mean()
```

```python
tetouan_grouped
```

```python
Power_Consumption_Zone1=tetouan_grouped["Zone 1 Power Consumption"]
Power_Consumption_Zone1
```

```python
Power_Consumption_Zone2=tetouan_grouped['Zone 2  Power Consumption']
Power_Consumption_Zone2
```

```python
Power_Consumption_Zone3=tetouan_grouped['Zone 3  Power Consumption']
Power_Consumption_Zone3
```

```python
%matplotlib inline
plt.plot(Power_Consumption_Zone1,'v--r',label='Zone 1  Power Consumption')
plt.plot(Power_Consumption_Zone2,label='Zone 2  Power Consumption')
plt.plot(Power_Consumption_Zone3,label='Zone 3  Power Consumption')
plt.legend()
plt.grid()
plt.show()
```

```python
%matplotlib inline
plt.plot(tetouan_grouped['Temperature'],'v--r',label='Temperature')
plt.plot(tetouan_grouped['Humidity'],label='Humidity')
plt.plot(tetouan_grouped['Wind Speed'],label='Wind Speed')
plt.legend()
plt.grid()
plt.show()
```

##Visualizations

```python
# Create numeric values for the x-axis
x = range(len(tetouan_grouped.index))

# Create two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot power consumption data for each zone in the first subplot
ax1.plot(x, Power_Consumption_Zone1, marker='o', label='Zone 1 Power Consumption')
ax1.plot(x, Power_Consumption_Zone2, marker='o', label='Zone 2 Power Consumption')
ax1.plot(x, Power_Consumption_Zone3, marker='o', label='Zone 3 Power Consumption')
ax1.set_xlabel('Time (Numeric Values)')
ax1.set_ylabel('Power Consumption')
ax1.set_xticks(x)
ax1.set_xticklabels([d.date() for d in tetouan_grouped.index], rotation=45)
ax1.legend(loc='upper left')
ax1.set_title('Power Consumption in Three Zones')

# Plot temperature, humidity, and wind speed in the second subplot
ax2.plot(x, tetouan_grouped['Temperature'], 'r', label='Temperature')
ax2.plot(x, tetouan_grouped['Humidity'], 'g', label='Humidity')
ax2.plot(x, tetouan_grouped['Wind Speed'], 'b', label='Wind Speed')
ax2.set_xlabel('Time (Numeric Values)')
ax2.set_ylabel('Temperature, Humidity, Wind Speed')
ax2.legend(loc='upper left')
ax2.set_title('Environmental Data')

# Add labels and titles
plt.suptitle('Power Consumption and Environmental Data')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract the data for "general diffuse flows" and the power consumption of each zone
general_diffuse_flows = tetouan['general diffuse flows']
zone1_power_consumption = tetouan['Zone 1 Power Consumption']
zone2_power_consumption = tetouan['Zone 2  Power Consumption']
zone3_power_consumption = tetouan['Zone 3  Power Consumption']

# Create scatter plots for each zone's power consumption
plt.figure(figsize=(12, 4))

# Scatter plot for Zone 1 Power Consumption
plt.subplot(131)
plt.scatter(general_diffuse_flows, zone1_power_consumption, alpha=0.5)
plt.title='Zone 1 Power Consumption vs. General Diffuse Flows'
plt.xlabel='General Diffuse Flows'
plt.ylabel='Zone 1 Power Consumption'
plt.grid()

# Scatter plot for Zone 2 Power Consumption
plt.subplot(132)
plt.scatter(general_diffuse_flows, zone2_power_consumption, alpha=0.5)
plt.title='Zone 2 Power Consumption vs. General Diffuse Flows'
plt.xlabel='General Diffuse Flows'
plt.ylabel='Zone 2 Power Consumption'
plt.grid()

# Scatter plot for Zone 3 Power Consumption
plt.subplot(133)
plt.scatter(general_diffuse_flows, zone3_power_consumption, alpha=0.5)
plt.title='Zone 3 Power Consumption vs. General Diffuse Flows'
plt.xlabel='General Diffuse Flows'
plt.ylabel='Zone 3 Power Consumption'
plt.grid()

plt.tight_layout()
plt.show()
```

#Finding weekday

```python
tetouan["weekday"]=tetouan['DateTime'].dt.day_name()
```

```python
tetouan
```

```python
tetouan_monday=tetouan[tetouan['weekday']=="Monday"]
```

```python
tetouan_monday
```

```python
tetouan_monday.index
```

```python
tetouan_monday_resampled = tetouan_monday.groupby(tetouan_monday.index.date).mean()
```

```python
tetouan_monday_resampled
```

```python
tetouan_monday_resampled.shape
```

```python
tetouan_monday_resampled.columns
```

```python
# Create numeric values for the x-axis
x = range(len(tetouan_monday_resampled.index))

# Create two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Format the x-axis labels to display only the date
date_format = mdates.DateFormatter("%Y-%m-%d")

# Plot power consumption data for each zone in the first subplot
ax1.plot(x, tetouan_monday_resampled['Zone 1 Power Consumption'], marker='o', label='Zone 1 Power Consumption')
ax1.plot(x, tetouan_monday_resampled['Zone 2  Power Consumption'], marker='o', label='Zone 2 Power Consumption')
ax1.plot(x, tetouan_monday_resampled['Zone 3  Power Consumption'], marker='o', label='Zone 3 Power Consumption')
ax1.set_xlabel('Time (Date)')
ax1.set_ylabel('Power Consumption')
ax1.set_xticks(x)
ax1.xaxis.set_major_formatter(date_format)  # Format x-axis labels
ax1.set_xticklabels([d.strftime("%Y-%m-%d") for d in tetouan_monday_resampled.index], rotation=90)
ax1.legend(loc='upper left')
ax1.set_title('Power Consumption in Three Zones')

# Plot temperature, humidity, and wind speed in the second subplot
ax2.plot(x, tetouan_monday_resampled['Temperature'], 'r', label='Temperature')
ax2.plot(x, tetouan_monday_resampled['Humidity'], 'g', label='Humidity')
ax2.plot(x, tetouan_monday_resampled['Wind Speed'], 'b', label='Wind Speed')
ax2.set_xlabel('Time (Date)')  # Update the x-axis label
ax2.set_ylabel('Temperature, Humidity, Wind Speed')
ax2.set_xticks(x)
ax2.xaxis.set_major_formatter(date_format)  # Format x-axis labels
ax2.set_xticklabels([d.strftime("%Y-%m-%d") for d in tetouan_monday_resampled.index], rotation=90)
ax2.legend(loc='upper left')
ax2.set_title('Environmental Data')

# Add labels and titles
plt.suptitle('Power Consumption and Environmental Data')
plt.tight_layout()
plt.show()
```

#Machine learning model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, accuracy_score
import matplotlib.pyplot as plt

# Assuming you have already loaded your data and created the 'tetouan_monday_resampled' DataFrame
# ...

# Extract features (X) and target variables (y)
X = tetouan_monday_resampled[['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows']]
y_zone1 = tetouan_monday_resampled['Zone 1 Power Consumption']
y_zone2 = tetouan_monday_resampled['Zone 2  Power Consumption']
y_zone3 = tetouan_monday_resampled['Zone 3  Power Consumption']

# Split the data into training and testing sets (80% training, 20% testing)
X_train_zone1, X_test_zone1, y_train_zone1, y_test_zone1 = train_test_split(X, y_zone1, test_size=0.2, random_state=42)
X_train_zone2, X_test_zone2, y_train_zone2, y_test_zone2 = train_test_split(X, y_zone2, test_size=0.2, random_state=42)
X_train_zone3, X_test_zone3, y_train_zone3, y_test_zone3 = train_test_split(X, y_zone3, test_size=0.2, random_state=42)

# Choose a machine learning model (Random Forest Regressor in this case)
model_zone1 = RandomForestRegressor(n_estimators=100, random_state=42)
model_zone2 = RandomForestRegressor(n_estimators=100, random_state=42)
model_zone3 = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
model_zone1.fit(X_train_zone1, y_train_zone1)
model_zone2.fit(X_train_zone2, y_train_zone2)
model_zone3.fit(X_train_zone3, y_train_zone3)

# Make predictions on the testing data
y_zone1_pred = model_zone1.predict(X_test_zone1)
y_zone2_pred = model_zone2.predict(X_test_zone2)
y_zone3_pred = model_zone3.predict(X_test_zone3)

# Evaluate the model using Mean Squared Error (MSE)
mse_zone1 = mean_squared_error(y_test_zone1, y_zone1_pred)
mse_zone2 = mean_squared_error(y_test_zone2, y_zone2_pred)
mse_zone3 = mean_squared_error(y_test_zone3, y_zone3_pred)

print(f'Mean Squared Error (Zone 1): {mse_zone1}')
print(f'Mean Squared Error (Zone 2): {mse_zone2}')
print(f'Mean Squared Error (Zone 3): {mse_zone3}')
```
