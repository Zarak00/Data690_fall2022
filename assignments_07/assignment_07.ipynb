```python
import pandas as pd
import plotly.express as px 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape
```

```python
df.sample(5)
```

```python
my_countries = ["ARE", "SWE", "IRQ"]

df2 = df[df["Country Code"].isin(my_countries)]

df2.shape
```

```python
df2["Country Name"].unique()
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df3 = df2.query('indicator == "SP.POP.TOTL" and Year == 2021')

df3
```

# Ploting Bar Chart

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2021 Population of UAE, Sweden, and Iraq", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

# Plot Pie Chart

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title="2021 Population of UAE, Sweden, and Iraq", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
df4 = df2.query('indicator == "SP.POP.TOTL"')

df4.head()
```

# Plotting Line Chart

```python
fig = px.line(
    df4,
    x="Year",
    y="value", 
    title="Population of UAE, Sweden, and Iraq over time", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
df["indicator"].unique()
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df5 = df[df["indicator"].isin(my_indicators)]

df5.shape
```

```python
df5.sample(10)
```

```python
df5 = df5[["Year",	"value",	"indicator",	"Country Code",	"Country Name"]]
          
df5.sample(5)
```

```python
df6 = pd.pivot_table(df5, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df6.head()
```

```python
df6.dropna(inplace=True)

df6.head()
```

```python
fig = px.scatter(
    df6.query('Year == 2020'),
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
